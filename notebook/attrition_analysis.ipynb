PHASE 1: DATA UNDERSTANDING

1.Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

2.Load Dataset

```python
df = pd.read_csv("D:/HR_Workforce_Attrition_Analysis/data/Professional_IBM_HR_Attrition_100K.csv")
```

6.Create Salary Band

```python
df['SalaryBand'] = pd.cut(
    df['MonthlyIncome'],
    bins=[0,5000,10000,50000],
    labels=['Low','Medium','High']
)
print(df['SalaryBand'].head())

print(df['SalaryBand'].value_counts())
salary_counts = df['SalaryBand'].value_counts().sort_index() 
salary_counts.plot( kind='bar', figsize=(6,4) ) 
plt.title('Employee Distribution by Salary Band') 
plt.xlabel('Salary Band')
plt.ylabel('Number of Employees') 
plt.savefig("../images/employess_vs_salarybond.png")
plt.show()
```

7.Create Experience Group

```python
df['ExperienceGroup'] = pd.cut(
    df['YearsAtCompany'],
    bins=[0,3,10,50],
    labels=['Fresher','Mid-Level','Senior']
)
print(df['ExperienceGroup'].head())
print(df['ExperienceGroup'].value_counts())
sns.countplot(x='ExperienceGroup', data=df)
plt.title('Experience Group Distribution')
plt.savefig("../images/experience_group_distribution.png")
plt.show()
```

8.Create Risk Level

```python
df['RiskLevel'] = 'Low Risk'
df.loc[
(df['JobSatisfaction'] <= 2) &
(df['WorkLifeBalance'] <= 2) &
(df['OverTime'] == 'Yes'),
'RiskLevel'
] = 'High Risk'


print(df['RiskLevel'].head())
print(df['RiskLevel'].value_counts())

sns.countplot(x='RiskLevel', data=df)
plt.title('Risk Level Distribution')
plt.savefig("../images/risk_level_analysis.png")
plt.show()
```

```python
print(pd.crosstab(
df['Department'],
df['Attrition']
))
sns.countplot(
x='Department',
hue='Attrition',
data=df
)
plt.xticks(rotation=45)
plt.savefig("../images/department_vs_attrition.png")
plt.show()
```

```python
print(pd.crosstab(
    df['SalaryBand'],
    df['Attrition']
))

sns.countplot(
    x='SalaryBand',
    hue='Attrition',
    data=df
)

plt.title('Salary Band vs Attrition')
plt.xlabel('Salary Band')
plt.ylabel('Employee Count')
plt.savefig("../images/salary_vs_attrition.png")
plt.show()
```

4.Overtime vs Attrition

```python
print(pd.crosstab(
    df['OverTime'],
    df['Attrition']
))

sns.countplot(
x='OverTime',
hue='Attrition',
data=df
)
plt.savefig("../images/overtime_vs_attrition.png")
plt.show()
```

5.Job Satisfaction vs Attrition

```python
print(pd.crosstab(
    df['JobSatisfaction'],
    df['Attrition']
))
sns.countplot(
x='JobSatisfaction',
hue='Attrition',
data=df
)
plt.savefig("../images/job_satisfication.png")
plt.show()
```

6.Work-Life Balance vs Attrition

```python
print(pd.crosstab(
    df['WorkLifeBalance'],
    df['Attrition']
))

sns.countplot(
x='WorkLifeBalance',
hue='Attrition',
data=df
)
plt.savefig("../images/work_life_balance.png")
plt.show()
```

7.KPI Cards

```python
# KPI Metrics

total_employees = len(df)

attrition_count = len(
    df[df['Attrition'] == 'Yes']
)

attrition_rate = (
    attrition_count / total_employees
) * 100

avg_salary = df['MonthlyIncome'].mean()

print("Total Employees :", total_employees)
print("Attrition Count :", attrition_count)
print("Attrition Rate  :", round(attrition_rate,2), "%")
print("Average Salary  :", round(avg_salary,2))
```

8.Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')
plt.savefig("../images/Correlation_heatmap.png")
plt.show()
```

9.Trend Analysis (Experience vs Attrition)

```python
# Create Experience Groups
df['ExperienceGroup'] = pd.cut(
    df['TotalWorkingYears'],
    bins=[0,5,10,20,40],
    labels=['0-5','5-10','10-20','20+']
)

trend = pd.crosstab(
    df['ExperienceGroup'],
    df['Attrition']
)

print(trend)

trend.plot(
    kind='line',
    marker='o',
    figsize=(8,5)
)

plt.title('Experience Group vs Attrition')
plt.ylabel('Employee Count')
plt.savefig("../images/experience_group_vs_attrition.png")
plt.show()
```

10.Funnel Analysis

```python
funnel_df = pd.DataFrame({
    'Stage':[
        'Total Employees',
        'Overtime Employees',
        'Low Satisfaction',
        'Attrition'
    ],
    'Count':[
        len(df),
        len(df[df['OverTime']=='Yes']),
        len(df[df['JobSatisfaction']<=2]),
        len(df[df['Attrition']=='Yes'])
    ]
})

print(funnel_df)

plt.figure(figsize=(8,5))

sns.barplot(
    x='Count',
    y='Stage',
    data=funnel_df
)

plt.title('Attrition Funnel')
plt.savefig("../images/attrition_funnel.png")
plt.show()
```

PHASE 4: EMPLOYEE SEGMENTATION

1.High-Risk Employees

```python
high_risk = df[df['RiskLevel'] == 'High Risk']

print("High Risk Employees:", len(high_risk))

print(high_risk['Department'].value_counts())

print(high_risk['Attrition'].value_counts())
```

2.High Performers

```python
high_performers = df[df['PerformanceRating'] >= 4]

print("High Performers:", len(high_performers))

print(high_performers['Attrition'].value_counts())

print(high_performers['Department'].value_counts())
```

3.Long-Term Employees

```python
long_term = df[df['YearsAtCompany'] > 10]

print("Long-Term Employees:", len(long_term))

print(long_term['Attrition'].value_counts())

print(long_term['JobSatisfaction'].mean())
```

4.Overtime Employees

```python
overtime_emp = df[df['OverTime'] == 'Yes']

print("Overtime Employees:", len(overtime_emp))

print(overtime_emp['Attrition'].value_counts())

print(overtime_emp['JobSatisfaction'].mean())
```

5.Dissatisfied Employees

```python
dissatisfied = df[df['JobSatisfaction'] <= 2]

print("Dissatisfied Employees:", len(dissatisfied))

print(dissatisfied['Attrition'].value_counts())

print(dissatisfied['OverTime'].value_counts())
```

Behavioral Difference Analysis

```python
segmentation = pd.DataFrame({
    'Group': [
        'High Risk',
        'High Performers',
        'Long-Term',
        'Overtime',
        'Dissatisfied'
    ],
    'Count': [
        len(high_risk),
        len(high_performers),
        len(long_term),
        len(overtime_emp),
        len(dissatisfied)
    ]
})

print(segmentation)

sns.barplot(
    x='Group',
    y='Count',
    data=segmentation
)

plt.title('Employee Segmentation')
plt.savefig("../images/employee_segmentation.png")
plt.xticks(rotation=45)
plt.show()
```
